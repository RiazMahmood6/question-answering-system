```python
from datasets import Dataset
# Ensure the required columns exist
assert set(["Question", "Answer"]).issubset(df.columns), "Dataset must have 'Question' and 'Answer' columns."

# Convert pandas DataFrame to Hugging Face Dataset
dataset = Dataset.from_pandas(df)

# Split into training and validation sets
data_split = dataset.train_test_split(test_size=0.2)
train_data = data_split['train']
val_data = data_split['test']
```

```python
def preprocess_data(examples):
    # Use the Question as input and Answer as target
    inputs = examples["Question"]
    targets = examples["Answer"]

    # Tokenize inputs and targets
    model_inputs = tokenizer(inputs, max_length=512, truncation=True)
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=128, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Tokenize the training and validation datasets
tokenized_train_data = train_data.map(preprocess_data, batched=True)
tokenized_val_data = val_data.map(preprocess_data, batched=True)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./finetuned_llama",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    num_train_epochs=3,
    save_steps=10,
    save_total_limit=2,
    logging_dir='./logs',
    fp16=True,  # Enable mixed precision for faster training
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_data,
    eval_dataset=tokenized_val_data,
    tokenizer=tokenizer,
)

trainer.train()
```

```python
# Save the model
model.save_pretrained("./finetuned_llama")
tokenizer.save_pretrained("./finetuned_llama")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model = AutoModelForCausalLM.from_pretrained("./finetuned_llama", device_map="auto")
tokenizer = AutoTokenizer.from_pretrained("./finetuned_llama")
```

```python
import gradio as gr

def answer_question(question):
    input_ids = tokenizer(question, return_tensors="pt").input_ids
    output = model.generate(input_ids, max_length=256)
    return tokenizer.decode(output[0], skip_special_tokens=True)

interface = gr.Interface(
    fn=answer_question,
    inputs="text",
    outputs="text",
    examples=[
        ["When did the Internet start in the United States?"],
        ["যুক্তরাষ্ট্রে ইন্টারনেট চালু হয় কবে"]
    ]
)

interface.launch()
```
